```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def van_der_pol(t, y, mu):
    x, dxdt = y
    dydt = [dxdt, mu * (1 - x**2) * dxdt - x]
    return dydt

mu = 1.0 
t_span = (0, 50)
y0 = [2, 0]

sol = solve_ivp(van_der_pol, t_span, y0, args=(mu,), t_eval=np.linspace(*t_span, 1000))

x = sol.y[0]
dxdt = sol.y[1]
t = sol.t

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, dxdt, t, color='b', lw=1.5)

ax.set_xlabel('x(t)')
ax.set_ylabel('dx/dt')
ax.set_zlabel('Time t')

plt.show()
```
